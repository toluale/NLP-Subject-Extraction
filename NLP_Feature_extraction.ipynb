```python
import feedparser
```

```python
url = 'https://pubmed.ncbi.nlm.nih.gov/rss/search/103QIAO1GYXhilW2OQIE0r6GfLU4jABmo10pwAnWaDIocp_gPZhttps/?limit=710&utm_campaign=pubmed-2&fc=20220531141644.xml'
```

```python
data = feedparser.parse(url)
```

```python
df_1 = {}
for article in data.entries:
  df_1.setdefault('Title', [])
  df_1.setdefault('Summary', [])
  df_1.setdefault('Authors', [])
  df_1['Title'].append(article.title)
  df_1['Summary'].append(article.summary)
  df_1['Authors'].append(article.authors)
```

```python
df_1
```

```python
df_2 = pd.DataFrame(df_1)
df_2
```

```python
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus =[]
heading = []
wnet = WordNetLemmatizer()

for i in range(len(df_2)):
  text = re.sub('[^a-zA-Z]', ' ', df_2['Summary'][i]).lower().split()
  text = [wnet.lemmatize(word) for word in text if not word in set(stopwords.words('english'))]
  text = ' '.join(text)
  corpus.append(text)

  head = re.sub('[^a-zA-Z]', ' ', df_2['Title'][i]).lower().split()
  head = [wnet.lemmatize(word) for word in head if not word in set(stopwords.words('english'))]
  head = ' '.join(head)
  heading.append(head)
```

```python
corpus
```

```python
heading
```

```python
df_3 = pd.DataFrame(corpus)
df_4 = pd.DataFrame(heading)
df_3.columns = ['summary']
df_4.columns = ['title']
new_data= df_4.join(df_3)
```
